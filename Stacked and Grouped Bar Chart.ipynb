```python
# -*- coding: utf-8 -*-
"""

@author: Roelof
"""

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("Chile data.csv")
df.rename(columns={"Povery Rate/Under $5.50 a day %":"Poverty Rate under $5.50 a day"},inplace=True)
df_top6=df.tail(10)
df_top6.plot.bar(x="Year",y=["Inflation rate","Poverty Rate under $5.50 a day"],stacked=True,rot=0);

plt.title("Inflation Rate/Poverty Rate between 2011-2020")
plt.xlabel("Years")
plt.ylabel("Rate")
plt.show()


import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

df=pd.read_csv("Chile data.csv")
df.rename(columns={"Povery Rate/Under $5.50 a day %": "Poverty Rate Under $5.50 a day"},inplace=True)
df_top6=df.tail(10)
df_top6.plot.bar(x="Year",y=["Inflation rate","Poverty Rate Under $5.50 a day"],rot=0);

plt.title("Inflation rate/Poverty Rate between 2011-2020")
plt.xlabel("Years")
plt.ylabel("Rate")
plt.show()


import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt





```
