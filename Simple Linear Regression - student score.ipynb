```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

df=pd.read_csv("Student Score.csv")


#Grab the columns and reshape to 2d array from a 1 day array
X=df.iloc[:,3].values.reshape(-1,1)
y=df.iloc[:,1].values.reshape(-1,1) #-1 means calculate dimensions of row but have 1 column

#Create object for class and perform Linear Regression
regressor=LinearRegression()
regressor.fit(X,y)

#Make Predictions- can also do (...fit(..).predict..)
y_pred=regressor.predict(X)


#Visualise data 
plt.scatter(X,y, s=14)
plt.title("Linear Regression - How House hold income affects exam scores")
plt.xlabel("Average Household Income")
plt.ylabel("Average exam scores")


#Predicted Values/Regression Line
plt.plot(X,y_pred,color="red",ls='solid',linewidth=1.0)
plt.show()

#Numerical Evaluation(Root mean squared error/ R-Squared)
rmse=mean_squared_error(y, y_pred)
r2=r2_score(y, y_pred)

#Print the Evaluation Values
print('Slope:',regressor.coef_)
print('Intercept:',regressor.intercept_)
print('Root mean squared error:',rmse)
print('R-Squared:',r2)
```
